```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
import numpy as np
import pandas as pd
from pathlib import Path
from trading_signal import signal_gen_2, signal_gen_1
from backtest import Timing_backtest_day
import warnings
warnings.filterwarnings('ignore')
import talib as ta
PATH1 = Path.cwd()
```

```python
subject = '沪深300期货主力日数据.xlsx'

start = '2015-07-01'
end = '2021-06-30'
backtest = Timing_backtest_day(signal_gen_1, subject, omit_len=20, interval=1)

close, signal, net, pnl1 = backtest.get_signal_net(start, end)
signals, per_pnl, performance = backtest.get_performance(signal, net, pnl1)
H_line, M_line, L_line = ta.BBANDS(close, timeperiod=20, nbdevup=2, nbdevdn=2, matype=0)

plt.figure(figsize = (20,8))
plt.plot(pnl1)
plt.show()

plt.figure(figsize = (20,8))
plt.plot(close, lw = 0.8, label = '收盘价')
plt.plot(H_line, lw = 0.8, label = '上轨')
plt.plot(M_line, lw = 0.8, label = '中轨')
plt.plot(L_line, lw = 0.8, label = '下轨')
plt.scatter(signal.index[signal == 1], close.loc[signal == 1].values, c = 'red', marker = '^', s = 30)
plt.scatter(signal.index[signal == -1], close.loc[signal == -1].values, c = 'green', marker = 'v', s = 30)
plt.scatter(signal.index[signal == 0], close.loc[signal == 0].values, c = 'black', s = 30)
plt.legend()
plt.show()

performance
```

```python
plt.figure(figsize = (20,8))
plt.plot(close.loc['20180701':'20190630'], lw = 0.8, label = '收盘价')
plt.plot(H_line.loc['20180701':'20190630'], lw = 0.8, label = '上轨')
plt.plot(M_line.loc['20180701':'20190630'], lw = 0.8, label = '中轨')
plt.plot(L_line.loc['20180701':'20190630'], lw = 0.8, label = '下轨')
plt.scatter(signal.loc['20180701':'20190630'].index[signal.loc['20180701':'20190630'] == 1], close.loc['20180701':'20190630'].loc[signal.loc['20180701':'20190630'] == 1].values, c = 'red', marker = '^', s = 100)
plt.scatter(signal.loc['20180701':'20190630'].index[signal.loc['20180701':'20190630'] == -1], close.loc['20180701':'20190630'].loc[signal.loc['20180701':'20190630'] == -1].values, c = 'green', marker = 'v', s = 100)
plt.scatter(signal.loc['20180701':'20190630'].index[signal.loc['20180701':'20190630'] == 0], close.loc['20180701':'20190630'].loc[signal.loc['20180701':'20190630'] == 0].values, c = 'black', s = 100)
plt.legend()
plt.show()
```
